```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/world-population-data-1960-2020/world_pop.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
year = [1960, 1970, 1980, 1990, 2000, 2010, 2020]
china_pop = [667070000, 818315000, 981235000, 1141700000, 1262660000, 1339180000, 1444216100]
india_pop = [449480608, 553417254, 696783490, 870133751, 1056575549, 1234281173, 1399880093]
united_states_pop = [180671000, 205052000, 227225000, 249623000, 281422000, 309349000, 331449281]
indonesia_pop = [87792515, 114793981, 147490298, 179377536, 206264595, 239870937, 273523615]
pakistan_pop = [44908293, 58196054, 78621312, 107147937, 138083480, 176745364, 225200000]
plt.stackplot(year, china_pop, india_pop, united_states_pop, indonesia_pop, pakistan_pop, labels=['China', 'India', 'United States', 'Indonesia', 'Pakistan'])
plt.legend(loc='upper left')
plt.title('Population Trends of Top 5 Countries')
plt.xlabel('Year')
plt.ylabel('Population (In Billions)')
plt.show()
```

```python
top_10 = df.sort_values('year_2020', ascending=False).head(10)
plt.figure(figsize=(18, 6))
plt.bar(top_10['country'], top_10['year_2020'])
plt.xlabel('Country')
plt.ylabel('Population (In Billions)')
plt.title('Top 10 Most Populated Countries in 2020')
plt.show()
```

```python
country = 'United States'
plt.plot(df.loc[df['country'] == country].iloc[:, 1:].values.T, label=country)
plt.xlabel('Year')
plt.ylabel('Population (In Billions)')
plt.title(f'Population Growth over Time: {country}')
plt.legend()
plt.show()
```

```python
countries = ['India', 'China']
populations = [df.loc[df['country'] == 'India', 'year_1961'].iloc[0],
               df.loc[df['country'] == 'China', 'year_1961'].iloc[0]]
plt.figure(figsize=(8, 4))
plt.bar(countries, populations)
plt.title('Population comparison between India and China in 1961')
plt.xlabel('Country')
plt.ylabel('Population (in millions)')
plt.show()

countries = ['India', 'China']
populations = [df.loc[df['country'] == 'India', 'year_2020'].iloc[0],
               df.loc[df['country'] == 'China', 'year_2020'].iloc[0]]
plt.figure(figsize=(8, 4))
plt.bar(countries, populations)
plt.title('Population comparison between India and China in 2020')
plt.xlabel('Country')
plt.ylabel('Population (in millions)')
plt.show()
```
