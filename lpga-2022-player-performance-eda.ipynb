Performance statistics and Prize winnings for 158 LPGA players in 2022 season.

| Variable   | Description                                               |
|------------|-----------------------------------------------------------|
| Golfer     | Name of the LPGA player                                   |
| Nation     | Nationality of the player                                 |
| Region     | Geographic region the player represents                   |
| fairways   | Number of drives that landed on fairways on par 4 and 5    |
| fairAtt    | Total number of drives on par 4 and 5 holes                |
| fairPct    | Percentage of drives that landed on fairways              |
| totPutts   | Total number of putts made                                |
| totRounds  | Total number of rounds played                             |
| avePutts   | Average number of putts per round                         |
| greenReg   | Percentage of greens hit in regulation (Par-2)            |
| totPrize   | Total official prize winnings in dollars                  |
| events     | Number of tournaments played in the 2022 season           |
| driveDist  | Average drive distance on par 4 and 5 holes in yards      |
| sandSaves  | Number of holes with par or better when in sandtrap       |
| sandAtt    | Number of times in the sandtrap                           |
| sandPct    | Percentage of times a player saved par or better in sand  |

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/lpga-performance-statistics-for-2022/lpga2022.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.set_style('whitegrid')
sns.scatterplot(x='driveDist', y='greenReg', data=df, color='#2980B9', alpha=0.7)
plt.xlabel('Average Drive Distance (yards)', fontsize=12, fontweight='bold')
plt.ylabel('Green Regulation Percentage', fontsize=12, fontweight='bold')
plt.title('Drive Distance vs. Green Regulation Percentage (2022)', fontsize=14, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['avePutts'], bins=10, kde=True)
plt.xlabel('Average Putts per Round')
plt.ylabel('Count')
plt.title('Distribution of Average Putts per Round (2022)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='fairPct', data=df)
plt.xlabel('Region')
plt.ylabel('Fairway Percentage')
plt.title('Fairway Percentage by Region (2022)')
plt.show()
```

```python
plt.figure(figsize=(16, 28))
plt.barh(y=df['Golfer'], width=df['totPrize'])
plt.xlabel('Total Prize Winnings ($)')
plt.ylabel('Player')
plt.title('Total Prize Winnings for LPGA Players (2022)')
plt.show()
```
