```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/arrests-for-marijuana-possession/Arrests.csv')
df.head()
```

```python
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.barplot(data=df, x='year', y='checks', estimator=sum)
plt.title('Barplot of Average Checks by Year')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
sns.countplot(data=df, x='colour')
plt.title('Countplot of Race')
plt.show()
```

```python
gender_counts = df['sex'].value_counts()
sns.set(style='whitegrid', font_scale=1.2)
fig, ax = plt.subplots(figsize=(4, 4))
sns.barplot(x=gender_counts.index, y=gender_counts.values, edgecolor='black', palette=['tab:blue', 'tab:red'])
plt.title('Countplot by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=df, x='colour', y='age')
plt.title('Boxplot of Age by Race')
plt.show()
```

```python
sns.violinplot(data=df, x='sex', y='age')
plt.title('Violinplot of Age by Sex')
plt.show()
```
